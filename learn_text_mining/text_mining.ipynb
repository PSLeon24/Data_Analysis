# 가장 많이 사용된 단어 알아보기

## 1. 연설문 불러오기

```python
# 그래프 해상도 설정
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi' : '300'})
%config InlineBackend.figure_format = 'retina'
```

```python
moon = open('./data/speech_moon.txt', encoding = 'UTF-8').read()
moon
```

## 2. 불필요한 문자 제거하기

```python
import re # 문자 처리 패키지
moon = re.sub('[^가-힣]', ' ', moon) # [^가-힣]은 '한글이 아닌 모든 문자'를 의미하는 정규 표현식(regular expression)
moon
```

## 3. 명사 추출하기
- 형태소 중에도 명사를 보면 텍스트의 내용을 파악할 수 있기 때문에 텍스트에서 명사만 추출해 분석할 때가 많다.
- conlpy.tag.Hannanum()의 nouns()를 사용하면 문장에서 명사를 추출할 수 있다.

```python
# hannanum 만들기
import konlpy
hannanum = konlpy.tag.Hannanum()

# 연설문에서 명사 추출하기
nouns = hannanum.nouns(moon)
nouns
```

```python
# 데이터 프레임으로 변환
import pandas as pd
df_word = pd.DataFrame({'word': nouns})
df_word.head()
```

## 4. 단어 빈도표 만들기

```python
# 글자 수 추가
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
# 단어 빈도 구하기
df_word = df_word.groupby('word', as_index = False) \
                  .agg(n = ('word', 'count')) \
                  .sort_values('n', ascending = False)
df_word
```

```python
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
from wordcloud import WordCloud

wc = WordCloud(random_state = 1234,
               font_path = 'NanumGothic',
               width = 400,
               height = 400,
               background_color = 'white')
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

```python
import PIL
import numpy as np

icon = PIL.Image.open('./data/cloud.png')
img = PIL.Image.new('RGB', icon.size, (255, 255, 255))
img.paste(icon, icon)
img = np.array(img)
```

```python
wc = WordCloud(random_state = 1234,
               font_path = 'NanumGothic',
               width = 400,
               height = 400,
               background_color = 'white',
               mask = img)
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
```
