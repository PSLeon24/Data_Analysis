## 초과, 미만, 이상, 이하 조건 걸기

```python
# math가 50점을 초과한 모든 행을 추출
exam.query('math > 50')
```

```python
# 수학 점수가 50점 미만인 모든 행을 추출
exam.query('math < 50')
```

```python
# english가 50점을 이상인 모든 행을 추출
exam.query('english >= 50')
```

```python
# english가 80점을 이하인 모든 행을 추출
exam.query('english <= 80')
```

## 여러 조건을 충족하는 행 추출하기

```python
# 1반이면서 수학 점수가 50점 이상인 경우
exam.query('nclass == 1 & math >= 50')
```

```python
# 2반이면서 영어 점수가 80점 이상인 경우
exam.query('nclass == 2 & english >= 80')
```

## 여러 조건 중 하나 이상 충족하는 행 추출하기

```python
# 수학 점수가 90점 이상이거나 영어 점수가 90점 이상인 경우
exam.query('math >= 90 | english >= 90')
```

```python
# 영어 점수가 90점 미만이거나 과학 점수가 50점 미만인 경우
exam.query('english < 90 | science < 50')
```

## 혼자서 해보기 p.144

```python
mpg = pd.read_csv('../Data/mpg.csv')
mpg
```

```python
# Prob1 - 144p
displ_under_4 = mpg.query('displ <= 4')
displ_upper_5 = mpg.query('displ >= 5')
print(f"배기량 4이하: {displ_under_4['hwy'].mean():.2f} \n배기량 5이상: {displ_upper_5['hwy'].mean():.2f}")
print('답: 배기량이 4이하인 자동차')
```

```python
# Prob2 - 144p
audi = mpg.query('manufacturer == "audi"')
toyota = mpg.query('manufacturer == "toyota"')
print(f"아우디: {audi['cty'].mean():.2f}")
print(f"토요타: {toyota['cty'].mean():.2f}")
print('AUDI의 평균이 더 높음') if audi['cty'].mean() > toyota['cty'].mean() else print('TOYOTA의 평균이 더 높음') 
```

```python
# Prob3 - 144p
cfh_hwy_mean = mpg.query('manufacturer in ["chevrolet", "ford", "honda"]')
print(f"hwy 전체 평균: {cfh_hwy_mean['hwy'].mean():.2f}")
```

## 혼자서 해보기 - p.150

```python
# Prob1 - p.150
mpg = pd.read_csv('../Data/mpg.csv')
new_mpg = mpg[['category', 'cty']]
new_mpg.head()
```

```python
# Prob2 - p.150
suv = new_mpg.query('category == "suv"')['cty'].mean()
compact = new_mpg.query('category == "compact"')['cty'].mean()
print(f'suv:{suv:.2f} compact:{compact:.2f}')
print(f'suv의 도시 연비 평균이 {suv-compact:.2f}만큼 더 높습니다') if suv > compact else print(f'compact의 도시 연비 평균이 {compact-suv:.2f}만큼 더 높습니다')
```

## lambda 이용해 데이터 프레임명 줄여 쓰기
