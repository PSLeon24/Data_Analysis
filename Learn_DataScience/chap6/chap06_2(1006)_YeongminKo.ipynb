## 06-6 집단별로 요약하기

```python
# p.160 - 1(전체 요약 통계량 구하기)
# df.agg() ~ 요약 통계량 구하는 함수
exam.agg(mean_math = ('math', 'mean'))
```

```python
# p.160 - 2(집단별 요약 통계량 구하기)
# df.groupby() 범주별로 데이터를 분리 해주는 함수
exam.groupby('nclass') \
    .agg(mean_math = ('math', 'mean')) # \ 뒤에는 띄어쓰기나 다른 어떤 문자도 안 돼!
```

```python
# p.161 - 1(변수를 인덱스로 바꾸지 않기)
exam.groupby('nclass', as_index=False) \
    .agg(mean_math = ('math', 'mean'))
```

```python
# p.161 - 2(여러 요약 통계량 한 번에 구하기)
exam.groupby('nclass') \
    .agg(mean_math = ('math', 'mean'),
         sum_math = ('math', 'sum'),
         median_math = ('math', 'median'),
         n = ('nclass', 'count'))
```

```python
# p.162 - 1(모든 변수의 요약 통계량 한 번에 구하기)
exam.groupby('nclass').mean()
```

```python
exam.groupby('nclass').sum() # p.162 - 1(모든 변수의 요약 통계량 한 번에 구하기 추가 연습)
```

```python
# p.163 - 1(집단별로 다시 집단 나누기)
mpg.groupby(['manufacturer', 'drv']) \
    .agg(mean_cty = ('cty', 'mean'))
```

```python
# p.163 - 2(집단별로 다시 집단 나누기)
mpg.query('manufacturer == "audi"') \
    .groupby(['drv']) \
    .agg(n = ('drv', 'count'))
```

```python
# p.164 - 1(집단별로 다시 집단 나누기)
mpg.query('manufacturer == "chevrolet"') \
    .groupby(['drv']) \
    .agg(n = ('drv', 'count'))
```

```python
# p.164 - 2(value_counts()로 집단별 빈도 간단하게 구하기)
mpg.groupby('drv') \
    .agg(n = ('drv', 'count'))
```

```python
# p.164 - 3(value_counts()로 집단별 빈도 간단하게 구하기)
mpg['drv'].value_counts()
```

```python
# p.165 - 1(value_counts()로 집단별 빈도 간단하게 구하기) / 시리즈에는 query() 사용 불가
mpg['drv'].value_counts().query('n > 100')
```

```python
# p.166 - 1(pandas 함수 조합하기)
mpg.query('category == "suv"') \
    .assign(total = (mpg['hwy'] + mpg['cty']) / 2) \
    .groupby('manufacturer') \
    .agg(mean_tot = ('total', 'mean')) \
    .sort_values('mean_tot', ascending = False) \
    .head()
```

## 06-7 데이터 합치기 - p.167
- 가로로 합치기: merge()
- 세로로 합치기: concat()

```python
# p.168 - 1(가로로 합치기)
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id'      : [1, 2, 3, 4, 5],
                      'mideterm': [60, 80, 70, 90, 85]})
# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id'      : [1, 2, 3, 4, 5],
                      'final': [70, 83, 65, 95, 80]})
```

```python
test1 # test1 출력
```

```python
test2 # test2 출력
```

```python
# p.169 - 1(가로로 합치기, id 기준으로 합쳐서 total에 할당)
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

```python
total = pd.merge(test1, test2, how = 'right', on = 'id')
total
```

```python
# p.169 - 2(다른 데이터를 활용해 변수 추가하기)
name = pd.DataFrame({'nclass' : [1, 2, 3, 4, 5],
                     'teacher': ['kim', 'lee', 'park', 'choi', 'jung']})
name
```

```python
# p.170 - 1(nclass 기준으로 합쳐서 exam_new에 할당)
exam_new = pd.merge(exam, name, how = 'left', on = 'nclass')
exam_new
```

```python
# p.170 - 2(세로로 합치기)
# 학생 1~5번 시험 데이터 만들기
group_a = pd.DataFrame({'id'  : [1, 2, 3, 4, 5],
                        'test': [60, 80, 70, 90, 85]})
# 학생 6~10번 시험 데이터 만들기
group_b = pd.DataFrame({'id'  : [6, 7, 8, 9, 10],
                        'test': [70, 83, 65, 95, 80]})
```

```python
# p.171 - 1(group_a 출력)
group_a
```

```python
# p.171 - 2(group_b 출력)
group_b
```

```python
# p.171 - 3(데이터 합쳐서 group_all에 할당)
group_all = pd.concat([group_a, group_b])
group_all
```

```python
group_all = pd.concat([group_a, group_b], ignore_index=True)
group_all
```
