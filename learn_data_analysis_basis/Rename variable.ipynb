# 데이터 프레임의 변수명 바꾸기
- pandas의 df.rename() 함수를 이용

## 1. Create DataFrame

```python
import pandas as pd
df_raw = pd.DataFrame({'var1': [1, 2, 1],
                       'var2': [2, 3, 2]})
df_raw
```

## 2. Copy DataFrame
- df.copy() 함수를 이용
- df.copy()를 사용하는 이유
    - 데이터에 문제가 생겼을 때 원래대로 복구하기 위해서 사본을 수정해도 원본은 영향을 받지 않도록 하기 위해서 사용해야 함

```python
df_new = df_raw.copy()
df_new
```

## 3. Rename variable
- df.rename(columns = {'기존 변수명': '새 변수명'})

```python
df_new = df_new.rename(columns = 
                        {'var1': 'v1',
                        'var2': 'v2'})
df_new
```

## 4. mpg data를 이용해 분석 문제 해결하기
- 1. mpg 데이터를 불러와 복사본을 만들기
- 2. 복사본 데이터를 이용해 cty는 city로, hwy는 highway로 수정하기
- 3. 데이터 일부를 출력해 변수명이 바뀌었는지 확인해보기

```python
mpg_df = pd.read_csv('./data/mpg.csv')
mpg_df.head()
```

```python
mpg_copy = mpg_df.copy()
mpg_copy = mpg_copy.rename(columns = {'cty': 'city',
                      'hwy': 'highway'})
mpg_copy.head()
```
