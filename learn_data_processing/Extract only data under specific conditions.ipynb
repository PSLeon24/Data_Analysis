# 조건에 맞는 데이터만 추출하기
- pandas의 df.query() 함수를 이용

```python
import pandas as pd
exam = pd.read_csv('./data/exam.csv')
exam.head()
```

```python
exam.query('nclass == 1')
```

```python
exam.query('english == 98')
```

```python
exam.query('nclass != 3')
```

## 초과, 미만, 이상, 이하 조건 걸기

```python
# 수학 점수 50점을 초과한 경우
exam.query('math > 50')
```

```python
# 수학 점수 50점 미만인 경우
exam.query('math < 50')
```

```python
# 영어 점수 50점 이상인 경우
exam.query('english >= 50')
```

```python
# 영어 점수 80점 이하인 경우
exam.query('english <= 80')
```

## 여러 조건을 충족하는 행 추출하기
- '그리고(and)'를 의미하는 & 기호를 사용

```python
# 1반이면서 수학 점수가 50점 이상인 경우 추출하기
exam.query('nclass == 1 & math >= 50')
```

```python
# 2반이면서 영어 점수가 80점 이상인 경우 추출하기
exam.query('nclass == 2 & english >= 80')
```

## 여러 조겅 중 하나 이상 충족하는 행 추출하기
- '또는(or)'를 의미하는 | 기호를 사용

```python
# 수학 점수가 90점 이상이거나 영어 점수가 90점 이상인 경우
exam.query('math >= 90 | english >= 90')
```

```python
# 영어 점수가 90점 미만이거나 과학 점수가 50점 미만인 경우
exam.query('english < 90 | science < 50')
```

## 목록에 해당하는 행 추출하기
- 예를 들어 1, 3, 5반 학생만 추출하고 싶으면 | 기호를 사용하여 여러 조건을 나열하면 된다.
- in과 []를 이용해 조건 목록을 입력하면 더 간결하게 작성 가능하다.

```python
exam.query('nclass == 1 | nclass == 3 | nclass == 5')
```

```python
exam.query('nclass in [1, 3, 5]')
```

## 추출한 행으로 데이터 만들기

```python
# nclass가 1인 행을 추출해 nclass1에 할당
nclass1 = exam.query('nclass == 1')

# nclass가 2인 행을 추출해 nclass2에 할당
nclass2 = exam.query('nclass == 2')
```

```python
nclass1['math'].mean()
```

```python
nclass2['math'].mean()
```

## 문자 변수를 이용해 조건에 맞는 행 추출하기

```python
df = pd.DataFrame({'sex': ['F', 'M', 'F', 'M'],
                   'country': ['Korea', 'China', 'Japan', 'USA']})
df
```

```python
df.query('sex == "F" & country == "Korea"')
```

# 실습: mpg 데이터를 이용해 분석 문제를 해결하기
#### 1. 자동차 배기량에 따라 고속도로 연비가 다른지 알아보려고 한다.<br> displ(배기량)이 4 이하인 자동차와 5 이상인 자동차 중 어떤 자동차의 hwy(고속도로 연비) 평균이 더 높은지 알아보자.

```python
mpg = pd.read_csv('./data/mpg.csv')
mpg.head()
```

```python
mpg_4_b = mpg.query('displ <= 4')
mpg_4_b['hwy'].mean()
```

```python
mpg_5_up = mpg.query('displ >= 5')
mpg_5_up['hwy'].mean()
```

#### 답: displ(배기량)이 4 이하인 자동차

#### 2. 자동차 제조 회사에 따라 도시 연비가 어떻게 다른지 알아보려고 한다.<br> 'audi'와 'toyota' 중 어느 manufacturer(자동차 제조 회사)의 cty(도시 연비) 평균이 더 높은지 알아보자.

```python
mpg_audi = mpg.query('manufacturer == "audi"')
mpg_audi['cty'].mean()
```

```python
mpg_toyota = mpg.query('manufacturer == "toyota"')
mpg_toyota['cty'].mean()
```

#### 답: toyoya에서 생산된 차량

#### 3. 'chevrolet', 'ford', 'honda' 자동차의 고속도로 연비 평균을 알아보려고 한다.<br> 세 회사의 데이터를 추출한 다음 hwy 전체 평균을 구하여 보자.

```python
mpg_3_company = mpg.query('manufacturer in ["chevrolet", "ford", "honda"]')
mpg_3_company['hwy'].mean()
```
